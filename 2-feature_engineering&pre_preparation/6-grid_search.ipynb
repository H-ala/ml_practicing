# Grid Search
automatically checking different alphas...

```python
import pandas as pd
```

```python
df = pd.read_csv('Advertising.csv')
df.head()
```

```python
# Create X and y
X = df.drop('sales', axis=1)
y = df['sales']

# Train and Split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)

# Scale Data
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(X_train)
X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
```

```python
# Modeling
from sklearn.linear_model import ElasticNet
base_elastic_net_model = ElasticNet()

# writing parameters for our model
param_grid = {'alpha': [0.1, 1, 5, 10, 50, 100], 'l1_ratio': [0.1, 0.5, 0.7, 0.95, 0.99, 1]} # checking 36 cases automatically
from sklearn.model_selection import GridSearchCV
grid_model = GridSearchCV(estimator=base_elastic_net_model, param_grid=param_grid, scoring='neg_mean_squared_error', cv=5, verbose=0)
```

```python
# it automatically chooses the best parameters and creates the best model
grid_model.fit(X_train, y_train)
```

```python
print(grid_model.best_params_)
grid_model.best_estimator_
```

```python
pd.DataFrame(grid_model.cv_results_)
```

```python
y_final = grid_model.predict(X_test)
from sklearn.metrics import mean_squared_error
mean_squared_error(y_test, y_final)
```
