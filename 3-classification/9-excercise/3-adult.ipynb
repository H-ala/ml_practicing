```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('adult.csv')
print(df.shape)
df.head()
```

```python
# there are a lot of object dtypes that we must check
df.info()
```

```python
# there is a '?' value which indicates missing values that we must fix
df['workclass'].value_counts()
```

```python
# in this data '?' chars are showing the missing values for us 
df_1 = df[df['workclass']=='?']
print(df.shape[0], df_1.shape[0]) # since missing values are a few segment of our data we can remove them
df_1.head()
```

```python
# we are ommiting rows that have '?' in the 'workclass' column
df = df[df['workclass'] != '?']
print(df.shape)
df.head()
```

```python
# df_categorical.apply(lambda x: print(x), axis=0)
# df_categorical.apply(lambda x: print(x), axis=1)
```

```python
df_categorical = df.select_dtypes(include='object')
df_categorical.apply(lambda x: x=='?', axis=0).sum() # checking which columns have '?' value and how many
```

```python
# since there were few of them we will remove them like before
df = df[df['occupation'] != '?']
df = df[df['native.country'] != '?']
print(df.shape)
df.head()
```

```python
# we have no longer have any '?' value in our rows
df_categorical = df.select_dtypes(include='object')
df_categorical.apply(lambda x: x=='?', axis=0).sum()
```

```python
# using LabelEncoder to convert our strings to numbers
# LaberEncoder creates relation between the values of the column, so it's better to use OneHotEncoding(get_dummies for convertion)
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()

df_categorical = df_categorical.apply(le.fit_transform) # default axis is 0
df_categorical.head()
```

```python
# omitting string columns
df = df.drop(df_categorical.columns, axis=1)

# adding the converted columns to the df so that all our columns would be int 
df = pd.concat([df, df_categorical], axis=1)
df.head() 
```

```python
# all of them are int64
df.info()
```
