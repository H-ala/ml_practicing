```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd     
import seaborn as sns
```

```python
df = pd.read_csv('Artificial.csv')
print(df.shape)
df.head()
```

```python
cdf = df.iloc[:, [0, 1]].values
cdf[:5]
```

```python
# it is apparent that data points in some parts are much denser and have more density
# can we use mean shift for datasets like this? where as data points density are different and it doesn't have an common shape
plt.scatter(cdf[:, 0], cdf[:, 1], alpha=0.6)
```

```python
from sklearn.cluster import MeanShift, estimate_bandwidth
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline


bandwidth = estimate_bandwidth(cdf, quantile=0.05)

pipe = Pipeline([
    ('scaler', StandardScaler()),
    ('meanshift', MeanShift(bandwidth=bandwidth))
])

pipe.fit(cdf)

labels = pipe.named_steps['meanshift'].labels_
print(np.unique(labels))
print(bandwidth)


plt.scatter(cdf[:, 0], cdf[:, 1], c=labels.astype(float), alpha=0.6);

plt.title('Cluster results on the datasets')
plt.xlabel('Variable X')
plt.ylabel('Variable Y')
plt.show()

# with 3 clusters, it will looks like this which is the same as kmeans
```

```python
df = pd.read_csv('Spiral.csv')
print(df.shape)
df.head()
```
