```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
# our data is splitted so we must merge it
train_df = pd.merge(pd.read_csv('X_train(1).csv'), pd.read_csv('y_train(1).csv'))
test_df = pd.read_csv('X_test(1).csv')
print(train_df.shape)
print(test_df.shape)
```

```python
# in every data csv, ids are not important and must be dropped
train_df.head()
```

```python
# names of the school
print(train_df['school'].unique())

# Urab or Rural
print(train_df['address'].unique())

# Family Size
print(train_df['famsize'].unique())

# Parent Status
print(train_df['Pstatus'].unique())
```

```python
train_df.isnull().sum()
```

```python
train_df.info()
```

```python
df = train_df.drop('StudentID', axis=1)
```

```python
df.head()
```

```python
df = pd.get_dummies(df, drop_first=True, dtype=int)
df.head()
```

```python
# we have 18 different labels? nah. it's a regression problem and we must predict a number
df['G3'].nunique()
```

```python
X = df.drop('G3', axis=1)   
y = df['G3']
```

```python
# Train and Split

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
# we will use rf regressor

from sklearn.ensemble import RandomForestRegressor
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline

model = Pipeline([
    ('scaler', StandardScaler()),
    ('rfr', RandomForestRegressor(random_state=42))
])

model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
                       
from sklearn.metrics import r2_score
r2 = r2_score(y_test, y_pred)
print(r2)
```
