```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Advertising.csv')
df.head()
```

```python
df["total_spent"] = df['TV'] + df['radio'] + df['newspaper']
df.head()
```

```python
# plt.scatter(df['total_spent'], df['sales'])
sns.scatterplot(data=df, x='total_spent', y='sales')
plt.show()
```

```python
# if we got only one x seaborn can compute the line for us with ols technique
sns.regplot(data=df, x='total_spent', y='sales')
plt.show()
```

```python
X = df['total_spent']
y = df['sales']
```

```python
# numpy has a func named polyfit 
# it can return coefficients and y-intercept of a line for us 
np.polyfit(X, y, deg=1)
```

```python
potential_spent = np.linspace(0, 500, 100) # create 100 random-ish number for me from 0 to 500
potential_spent
```

```python
predicted_sales = 0.04868788 * potential_spent + 4.24302822
predicted_sales 
```

```python
sns.scatterplot(data=df, x='total_spent', y='sales')
plt.plot(potential_spent, predicted_sales, color='r')
plt.show()
```
